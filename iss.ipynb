```python
import requests
import time
data = []
for i in range(150):
    res = requests.get("http://api.open-notify.org/iss-now.json")
    json_data = res.json()
    data.append(json_data)
    time.sleep(10)
```

```python
data
```

```python
body = []
for record in data:
        row = []
        row.append(record['timestamp'])
        row.append(record['iss_position']['latitude'])
        row.append(record['iss_position']['longitude'])
        body.append(row)
```

```python
from matplotlib import pyplot as plt

data_rawr = list(zip(*body))

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('ISS Tracker')

x = data_rawr[2]
y = data_rawr[1]

plt.plot(x, y, color='#e0218a')
plt.show()
```
